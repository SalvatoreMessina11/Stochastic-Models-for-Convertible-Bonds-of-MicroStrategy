## setting

```python
# ===== libraries =====
from __future__ import annotations
import os, json
import numpy as np
import numba as nb
from dataclasses import dataclass
from datetime import datetime, timezone

# SciPy: norm + Sobol con fallback
try:
    from scipy.stats import norm
    from scipy.stats.qmc import Sobol
except Exception:
    from scipy.stats import norm
    from scipy.stats import qmc as _qmc
    Sobol = _qmc.Sobol

# ===== pre-caricamento meta =====
path = "mstr_panel_meta.json"
if os.path.exists(path):
    with open(path, "r") as f:
        meta = json.load(f)
else:
    meta = {}

S0 = float(meta.get("S0", 100.0))
r  = float(meta.get("r", 0.035))
spot_ts = meta.get("spot_ts")
if spot_ts:
    BASE_DATE = datetime.fromisoformat(spot_ts.replace("Z","+00:00"))
else:
    BASE_DATE = datetime(2025, 10, 24, 12, 55, 51, 480405, tzinfo=timezone.utc)

# alias usati piÃ¹ sotto
S0_meta = S0
r_meta  = r

print(f"S0: {S0}")
print(f"r: {r}")
print(f"BASE_DATE: {BASE_DATE.isoformat()}")

# ===== setup dopo pre-caricamento =====
N_UNIT = 100.0

def _yf(d: datetime, base: datetime = BASE_DATE) -> float:
    if d.tzinfo is None and base.tzinfo is not None:
        d = d.replace(tzinfo=base.tzinfo)
    return (d - base).days / 365.0

def _bond_spec(tipo: str):
    t = str(tipo).strip().upper()
    if t == "2029":
        Cr = 0.14872
        S_star = N_UNIT / Cr
        K_soft = 1.3 * S_star
        return dict(
            Cr=Cr, S_star=S_star, K_soft=K_soft,
            t_put=_yf(datetime(2028, 6, 1)),
            t_soft_a=_yf(datetime(2026, 12, 2)),
            t_soft_b=_yf(datetime(2029, 11, 29)),
            T=_yf(datetime(2029, 11, 29)),
            name="2029"
        )
    if t in {"2030", "2030B"}:
        Cr = 0.23072
        S_star = N_UNIT / Cr
        K_soft = 1.3 * S_star
        return dict(
            Cr=Cr, S_star=S_star, K_soft=K_soft,
            t_put=_yf(datetime(2028, 3, 1)),
            t_soft_a=_yf(datetime(2027, 3, 5)),
            t_soft_b=_yf(datetime(2030, 2, 1)),
            T=_yf(datetime(2030, 2, 27)),
            name="2030B"
        )
    raise ValueError("tipo deve essere '2029' o '2030B'")

@dataclass
class ConvertibleMCResult:
    price: float
    d1: float
    d2: float
    d3: float
    d4: float
    Sput: np.ndarray
    Vput: np.ndarray
    S_eq_N: float = np.nan
```

## nested montecarlo aligned (QMC) 

```python
# ===== nested montecarlo aligned (QMC) =====
@nb.njit(parallel=True, fastmath=True)
def _qmc_tail_vec_kernel(
    S_put_vec, mu, sigma, r,
    dt, t, tb_idx, k_soft, N, conv_ratio,
    Z
):
    n2 = S_put_vec.shape[0]
    M1, n_steps = Z.shape
    sqrt_dt = np.sqrt(dt)
    out = np.empty(n2)
    for i in nb.prange(n2):
        s0 = S_put_vec[i]
        acc = 0.0
        for m in range(M1):
            s = s0
            hit = False
            pv = 0.0
            for k in range(n_steps):
                s = s * np.exp(mu*dt[k] + sigma*sqrt_dt[k]*Z[m, k])
                tau = t[k+1]
                if tb_idx > 0 and (k+1) <= tb_idx and s > k_soft:
                    pv = np.exp(-r * tau) * (conv_ratio * s)
                    hit = True
                    break
            if not hit:
                conv_val = conv_ratio * s
                pv = np.exp(-r * t[-1]) * (N if conv_val < N else conv_val)
            acc += pv
        out[i] = acc / M1
    return out

def V_put_QMC_vec(dt_b, dt_T, S_put_vec, sigma, r, k_soft, M1, N, conv_ratio,
                  steps_per_year=252, seed=None):
    dt_b = float(max(0.0, dt_b)); dt_T = float(max(0.0, dt_T))
    S_put_vec = np.asarray(S_put_vec, dtype=np.float64)
    if dt_T == 0.0:
        return np.maximum(N, conv_ratio * S_put_vec)

    n_tail = max(1, int(np.ceil(dt_T * steps_per_year)))
    t_reg = np.linspace(0.0, dt_T, n_tail + 1, dtype=np.float64)
    t = np.unique(np.concatenate([t_reg, np.array([0.0, min(dt_b, dt_T), dt_T], dtype=np.float64)])).astype(np.float64)
    dt = np.diff(t).astype(np.float64)

    tb_idx = int(np.searchsorted(t, min(dt_b, dt_T), side="left"))
    if dt_b <= 0.0: tb_idx = 0

    d = dt.shape[0]
    rng = Sobol(d=d, scramble=True, seed=seed)
    if (M1 & (M1-1)) == 0 and M1 > 0:
        U = rng.random_base2(int(np.log2(M1)))
    else:
        U = rng.random(M1)
    U = np.clip(U, 1e-12, 1-1e-12)
    Z = norm.ppf(U).astype(np.float64)

    mu = float(r - 0.5 * sigma * sigma)
    return _qmc_tail_vec_kernel(S_put_vec, mu, float(sigma), float(r),
                                dt, t, int(tb_idx), float(k_soft), float(N), float(conv_ratio), Z)

@nb.njit(parallel=True, fastmath=True)
def _front_phase_kernel(S0, mu, sigma, r, K_soft, conv_ratio, dt_grid, t_grid, Z):
    N1, n_steps = Z.shape
    sqrt_dt = np.sqrt(dt_grid)
    pv_hit = np.zeros(N1, dtype=np.float64)
    S_put  = np.empty(N1, dtype=np.float64)
    for i in nb.prange(N1):
        s = S0
        hit = False
        pv = 0.0
        for k in range(n_steps):
            s = s * np.exp(mu*dt_grid[k] + sigma*sqrt_dt[k]*Z[i, k])
            tau_abs = t_grid[k+1]
            if s > K_soft:
                pv = np.exp(-r * tau_abs) * (conv_ratio * s)
                hit = True
                break
        if hit:
            pv_hit[i] = pv; S_put[i] = np.nan
        else:
            pv_hit[i] = 0.0; S_put[i] = s
    return pv_hit, S_put

def _single_path_classify_after_put(S0_vec, sigma, r, K_soft, dt_b, dt_T, steps_per_year, seed=None):
    S0_vec = np.asarray(S0_vec, dtype=np.float64); n = S0_vec.shape[0]
    if n == 0: return np.zeros(0, dtype=bool), np.zeros(0, dtype=bool)
    rng = np.random.default_rng(seed)
    n_tail = max(1, int(np.ceil(dt_T * steps_per_year)))
    t = np.linspace(0.0, dt_T, n_tail + 1); dt = np.diff(t).astype(np.float64)
    sqrt_dt = np.sqrt(dt)
    tb_time = min(dt_b, dt_T)
    tb_idx = int(np.searchsorted(t, tb_time, side="left"))
    if dt_b <= 0.0: tb_idx = 0
    mu = float(r - 0.5*sigma*sigma)
    hit_before_tb = np.zeros(n, dtype=bool); reaches_T = np.zeros(n, dtype=bool)
    for i in range(n):
        s = S0_vec[i]; Z = rng.standard_normal(dt.shape[0])
        for k in range(dt.shape[0]):
            s *= np.exp(mu*dt[k] + sigma*sqrt_dt[k]*Z[k])
            if tb_idx > 0 and (k+1) <= tb_idx and s > K_soft:
                hit_before_tb[i] = True; break
        else:
            reaches_T[i] = True
    return hit_before_tb, reaches_T

def _S_eq_nom_from_samples(S, V_cont_tput, Nom):
    S = np.asarray(S, dtype=float); V = np.asarray(V_cont_tput, dtype=float)
    if S.size == 0: return np.nan
    idx = np.argsort(S); S = S[idx]; V = V[idx]
    crosses = np.where((V[:-1] <= Nom) & (V[1:] >= Nom))[0]
    if crosses.size == 0: return np.nan
    roots = []
    for i in crosses:
        v0, v1 = V[i], V[i+1]; s0, s1 = S[i], S[i+1]
        if v1 == v0: continue
        w = (Nom - v0) / (v1 - v0); roots.append(s0 + w * (s1 - s0))
    return float(np.max(roots)) if roots else np.nan

def price_convertibile_soft_put_stats(S0, r, Nom, sigma, t0_val, t_a, t_put, t_b, T,
                                      K_spot, N1, M1, steps_per_year=252, seed=None, K_soft=None, Cr=1.0) -> ConvertibleMCResult:
    dt_put_sim  = float(max(0.0, t_put - t_a))
    dt_b        = float(max(0.0, t_b   - t_put))
    dt_T        = float(max(0.0, T     - t_put))
    dt_put_disc = float(max(0.0, t_put - t0_val))
    dt_a_disc   = float(max(0.0, t_a   - t0_val))
    K_soft_eff  = float(K_spot if K_soft is None else K_soft)

    n_front = max(1, int(np.ceil(max(dt_put_sim, 1e-12) * steps_per_year)))
    t_grid = np.linspace(0.0, dt_put_sim, n_front + 1, dtype=np.float64)
    dt_grid = np.diff(t_grid).astype(np.float64); d = dt_grid.shape[0]

    rng_front = Sobol(d=d, scramble=True, seed=seed)
    if (N1 & (N1-1)) == 0 and N1 > 0: U = rng_front.random_base2(int(np.log2(N1)))
    else: U = rng_front.random(N1)
    U = U[:N1]; U = np.clip(U, 1e-12, 1-1e-12)
    Z_front = norm.ppf(U).astype(np.float64)
    mu = float(r - 0.5*sigma*sigma)

    pv_hit_ta, S_put_all = _front_phase_kernel(float(S0), mu, float(sigma), float(r), float(K_soft_eff), float(Cr),
                                               dt_grid, t_grid, Z_front)

    mask_surv = ~np.isnan(S_put_all)
    d1_cnt = int((~mask_surv).sum()); d1 = d1_cnt / float(N1)
    media1_today = np.exp(-r * dt_a_disc) * (pv_hit_ta.sum() / float(N1))

    S_put_vec = S_put_all[mask_surv].astype(np.float64)
    if S_put_vec.size == 0:
        return ConvertibleMCResult(price=float(media1_today), d1=d1, d2=0.0, d3=0.0, d4=0.0,
                                   Sput=np.array([], dtype=float), Vput=np.array([], dtype=float), S_eq_N=np.nan)

    V_cont_tput = V_put_QMC_vec(dt_b, dt_T, S_put_vec, sigma, r, K_soft_eff, int(M1), Nom, Cr,
                                steps_per_year=steps_per_year, seed=None if seed is None else seed+1)

    V_ex_tput = np.where(Nom >= V_cont_tput, Nom, V_cont_tput)
    put_mask = (V_ex_tput == Nom); d2 = float(put_mask.sum()) / float(N1)

    V_ex_today = np.exp(-r * dt_put_disc) * V_ex_tput
    price_today = float(media1_today + V_ex_today.sum() / float(N1))

    S_cont = S_put_vec[~put_mask]
    if S_cont.size > 0 and (dt_T > 0.0):
        hit_before_tb, reaches_T = _single_path_classify_after_put(S_cont, sigma, r, K_soft_eff, dt_b, dt_T, steps_per_year,
                                                                   seed=None if seed is None else seed+2)
        d3_cnt = int(hit_before_tb.sum()); d4_cnt = int(reaches_T.sum())
    else:
        d3_cnt = 0; d4_cnt = 0
    d3 = d3_cnt / float(N1); d4 = d4_cnt / float(N1)

    S_eq_N = _S_eq_nom_from_samples(S_put_vec, V_cont_tput, float(Nom))
    idx = np.argsort(S_put_vec)
    return ConvertibleMCResult(price=price_today, d1=d1, d2=d2, d3=d3, d4=d4,
                               Sput=S_put_vec[idx], Vput=V_ex_tput[idx], S_eq_N=S_eq_N)

def BS_exotic(S0, sigma, r, N1, M1, tipo, steps_per_year=252, seed=None):
    spec   = _bond_spec(tipo)
    Cr     = float(spec["Cr"])
    t0_val = 0.0
    t_a    = float(spec["t_soft_a"])
    t_put  = float(spec["t_put"])
    t_b    = float(spec["t_soft_b"])
    T      = float(spec["T"])
    K_soft = float(spec["K_soft"])
    Nom    = float(N_UNIT)
    K_spot = K_soft

    res = price_convertibile_soft_put_stats(S0=S0, r=r, Nom=Nom, sigma=sigma,
                                            t0_val=t0_val, t_a=t_a, t_put=t_put, t_b=t_b, T=T,
                                            K_spot=K_spot, N1=int(N1), M1=int(M1),
                                            steps_per_year=steps_per_year, seed=seed, K_soft=K_soft, Cr=Cr)

    print(f"BS_exotic {spec['name']}: price (today) = {res.price:.6f}")
    print(f"K_soft: {K_soft:.6f}")
    print(f"d1/N1 (called before put): {res.d1:.6%}")
    print(f"d2/N1 (put at t_put):      {res.d2:.6%}")
    print(f"d3/N1 (called < t_b):      {res.d3:.6%}")
    print(f"d4/N1 (reach T):           {res.d4:.6%}")
    print(f"S_max such that cont(S) = {Nom:.0f}: {res.S_eq_N:.6f}")
    return res
```

# LSM


```python
# ===== lsm montecarlo =====
def _simulate_to_tput_LSMC(S0, r, sigma, K_soft, t_a, t_put, steps_per_year, N1, Cr, seed=None):
    dt_put_sim = float(max(0.0, t_put - t_a))
    n_steps = max(1, int(np.ceil(max(dt_put_sim, 1e-12) * steps_per_year)))
    t_grid = np.linspace(0.0, dt_put_sim, n_steps + 1, dtype=np.float64)
    dt = np.diff(t_grid).astype(np.float64); sqrt_dt = np.sqrt(dt)
    mu = float(r - 0.5 * sigma * sigma)

    rng = Sobol(d=n_steps, scramble=True, seed=seed)
    if (N1 & (N1-1)) == 0 and N1 > 0: U = rng.random_base2(int(np.log2(N1)))
    else: U = rng.random(N1)
    U = np.clip(U, 1e-12, 1-1e-12)
    Z = norm.ppf(U).astype(np.float64)

    pv_pre_ta = np.zeros(N1, dtype=np.float64); S_put_all = np.empty(N1, dtype=np.float64)
    for i in range(N1):
        s = S0; pv = 0.0; alive = True
        for k in range(n_steps):
            s *= np.exp(mu*dt[k] + sigma*sqrt_dt[k]*Z[i, k])
            tau_abs = t_grid[k+1]
            if s > K_soft:
                pv = np.exp(-r * tau_abs) * (Cr * s); alive = False; break
        if alive:
            pv_pre_ta[i] = 0.0; S_put_all[i] = s
        else:
            pv_pre_ta[i] = pv; S_put_all[i] = np.nan
    return dt_put_sim, pv_pre_ta, S_put_all

def _tail_one_path_Y_and_flags(S_put_vec, r, sigma, K_soft, Cr, Nom, t_put, t_b, T, steps_per_year, seed=None):
    n = S_put_vec.size
    if n == 0: return np.zeros(0), np.zeros(0, dtype=bool), np.zeros(0, dtype=bool)
    dt_b = float(max(0.0, t_b - t_put)); dt_T = float(max(0.0, T - t_put))
    n_tail = max(1, int(np.ceil(max(dt_T, 1e-12) * steps_per_year)))
    t = np.linspace(0.0, dt_T, n_tail + 1, dtype=np.float64)
    dt = np.diff(t).astype(np.float64); sqrt_dt = np.sqrt(dt)
    mu = float(r - 0.5 * sigma * sigma)
    tb_idx = int(np.searchsorted(t, min(dt_b, dt_T), side="left"))
    if dt_b <= 0.0: tb_idx = 0
    rng = np.random.default_rng(seed); Z = rng.standard_normal((n, n_tail))

    Y_pv_tput = np.empty(n, dtype=np.float64); hit_before_tb = np.zeros(n, dtype=bool); reaches_T = np.zeros(n, dtype=bool)
    for i in range(n):
        s = S_put_vec[i]; pv = None
        for k in range(n_tail):
            s *= np.exp(mu*dt[k] + sigma*sqrt_dt[k]*Z[i, k])
            tau_rel = t[k+1]
            if tb_idx > 0 and (k+1) <= tb_idx and s > K_soft:
                pv = np.exp(-r * tau_rel) * (Cr * s); hit_before_tb[i] = True; break
        if pv is None:
            pv = np.exp(-r * t[-1]) * max(Nom, Cr * s); reaches_T[i] = True
        Y_pv_tput[i] = pv
    return Y_pv_tput, hit_before_tb, reaches_T

def _fit_log_poly_continuation(S_put_vec, Y_pv_tput):
    L = np.log(S_put_vec)
    X = np.column_stack([np.ones_like(L), L, L**2, L**3])
    beta, _, _, _ = np.linalg.lstsq(X, Y_pv_tput, rcond=None)
    Chat = X @ beta
    return Chat, beta

def _max_S_eq_N(beta, Nom):
    b0, b1, b2, b3 = beta.tolist()
    coeffs = np.array([b3, b2, b1, b0 - Nom], dtype=float)
    k = 0
    while k < coeffs.size - 1 and abs(coeffs[k]) < 1e-14: k += 1
    coeffs = coeffs[k:]
    roots = np.roots(coeffs)
    real_L = roots[np.isclose(roots.imag, 0.0, atol=1e-8)].real
    if real_L.size == 0: return np.nan
    return float(np.exp(real_L.max()))

def price_convertibile_soft_put_stats_LSMC(S0, r, Nom, sigma, t0_val, t_a, t_put, t_b, T,
                                           K_soft, N1, steps_per_year=252, seed=None, Cr=1.0) -> ConvertibleMCResult:
    dt_put_sim, pv_pre_ta, S_put_all = _simulate_to_tput_LSMC(float(S0), float(r), float(sigma), float(K_soft),
                                                              float(t_a), float(t_put), int(steps_per_year), int(N1), float(Cr), seed=seed)
    dt_a_disc = float(max(0.0, t_a - t0_val))
    media1_today = np.exp(-r * dt_a_disc) * (pv_pre_ta.sum() / float(N1))

    mask_alive = ~np.isnan(S_put_all); d1_cnt = int((~mask_alive).sum()); d1 = d1_cnt / float(N1)
    S_put_vec = S_put_all[mask_alive].astype(np.float64)
    if S_put_vec.size == 0:
        return ConvertibleMCResult(price=float(media1_today), d1=d1, d2=0.0, d3=0.0, d4=0.0,
                                   Sput=np.array([], dtype=float), Vput=np.array([], dtype=float), S_eq_N=np.nan)

    Y_pv_tput, hit_before_tb, reaches_T = _tail_one_path_Y_and_flags(S_put_vec, float(r), float(sigma), float(K_soft), float(Cr), float(Nom),
                                                                     float(t_put), float(t_b), float(T), int(steps_per_year),
                                                                     seed=None if seed is None else seed + 11)
    Chat_pv_tput, beta = _fit_log_poly_continuation(S_put_vec, Y_pv_tput)
    V_ex_tput = np.where(float(Nom) >= Chat_pv_tput, float(Nom), Chat_pv_tput)
    put_mask = (V_ex_tput == float(Nom)); d2 = float(put_mask.sum()) / float(N1)

    dt_put_disc = float(max(0.0, t_put - t0_val))
    V_ex_today = np.exp(-r * dt_put_disc) * V_ex_tput
    price_today = float(media1_today + V_ex_today.sum() / float(N1))

    cont_mask = ~put_mask
    d3_cnt = int(hit_before_tb[cont_mask].sum()); d4_cnt = int(reaches_T[cont_mask].sum())
    d3 = d3_cnt / float(N1); d4 = d4_cnt / float(N1)
    S_eq_N = _max_S_eq_N(beta, float(Nom))

    idx = np.argsort(S_put_vec)
    return ConvertibleMCResult(price=price_today, d1=d1, d2=d2, d3=d3, d4=d4,
                               Sput=S_put_vec[idx], Vput=V_ex_tput[idx], S_eq_N=S_eq_N)

def BS_exotic_LSMC(S0, sigma, r, N1, tipo, steps_per_year=252, seed=None):
    spec = _bond_spec(tipo)
    Cr = float(spec["Cr"])
    t0_val = 0.0
    t_a    = float(spec["t_soft_a"])
    t_put  = float(spec["t_put"])
    t_b    = float(spec["t_soft_b"])
    T      = float(spec["T"])
    K_soft = float(spec["K_soft"])
    Nom    = float(N_UNIT)

    res = price_convertibile_soft_put_stats_LSMC(S0=S0, r=r, Nom=Nom, sigma=sigma,
                                                 t0_val=t0_val, t_a=t_a, t_put=t_put, t_b=t_b, T=T,
                                                 K_soft=K_soft, N1=int(N1),
                                                 steps_per_year=steps_per_year, seed=seed, Cr=Cr)

    print(f"BS_exotic {spec['name']} (LSMC): price (today) = {res.price:.6f}")
    print(f"K_soft: {K_soft:.6f}")
    print(f"d1/N1 (called before put): {res.d1:.6%}")
    print(f"d2/N1 (put at t_put):      {res.d2:.6%}")
    print(f"d3/N1 (called < t_b):      {res.d3:.6%}")
    print(f"d4/N1 (reach T):           {res.d4:.6%}")
    if res.Sput.size > 0:
        L = np.log(res.Sput)
        X = np.column_stack([np.ones_like(L), L, L**2, L**3])
        beta, _, _, _ = np.linalg.lstsq(X, res.Vput, rcond=None)
        b0, b1, b2, b3 = beta.tolist()
        print("Regression @ t_put on [1, logS, (logS)^2, (logS)^3]:")
        print(f"V_put â {b0:.6f} + ({b1:.6f})Â·logS + ({b2:.6f})Â·(logS)^2 + ({b3:.6f})Â·(logS)^3")
    else:
        print("Regression @ t_put non disponibile.")
    print(f"S_max such that cont(S) = {Nom:.0f}: {res.S_eq_N:.6f}")
    return res
```

## implied vol

```python
# === Implied-vol "smirk" da chain_data.csv ===
# Requisiti: numpy, pandas, scipy, matplotlib

import json, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import brentq

# ====== BlackâScholes ======
def _d1(S0, K, T, r, q, sigma):
    if sigma <= 0 or T <= 0:
        return np.inf
    return (np.log(S0/K) + (r - q + 0.5*sigma*sigma)*T) / (sigma*np.sqrt(T))

def bs_call(S0, K, T, r, q, sigma):
    if T <= 0:
        return max(S0 - K, 0.0)
    d1 = _d1(S0,K,T,r,q,sigma)
    d2 = d1 - sigma*np.sqrt(T)
    return float(S0*np.exp(-q*T)*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2))

def implied_vol_call(price, S0, K, T, r, q, lo=1e-6, hi=5.0):
    if T <= 0:
        return np.nan
    intrinsic = max(S0*np.exp(-q*T) - K*np.exp(-r*T), 0.0)
    if price < intrinsic:
        return np.nan
    f = lambda s: bs_call(S0,K,T,r,q,s) - price
    try:
        a, b = lo, hi
        fa, fb = f(a), f(b)
        if not np.isfinite(fa) or not np.isfinite(fb):
            return np.nan
        if fa*fb > 0:
            for b_try in (8.0, 12.0):
                fb = f(b_try)
                if fa*fb <= 0:
                    b = b_try
                    break
        return float(brentq(f, a, b, maxiter=200, xtol=1e-8))
    except Exception:
        return np.nan

# ====== Carica dati ======
# File prezzi
CHAIN_FILE = "chain_data.csv"          # <<-- file reale
# Meta con S0, r, q
with open("mstr_panel_meta.json","r") as f:
    meta = json.load(f)

S0 = float(meta.get("S0", 100.0))
r_default = float(meta.get("r", 0.0))
q = float(meta.get("q", 0.0))
ticker = meta.get("ticker","TICKER")

df = pd.read_csv(CHAIN_FILE)

# Supporta due formati:
# A) colonne: maturity, strike, price, rate  (screenshot)
# B) colonne: T, K, close, right            (vecchio formato)
if {"maturity","strike","price"}.issubset(df.columns):
    df = df.rename(columns={"maturity":"T","strike":"K","price":"close"})
    if "rate" in df.columns:
        df["r"] = df["rate"].astype(float)
    else:
        df["r"] = r_default
elif {"T","K","close"}.issubset(df.columns):
    df["r"] = r_default
else:
    raise ValueError("Colonne non riconosciute. Attese: (maturity,strike,price[,rate]) oppure (T,K,close).")

# Pulisci
df = df.dropna(subset=["K","T","close"])
df = df[(df["close"]>0) & np.isfinite(df["close"]) & np.isfinite(df["T"]) & np.isfinite(df["K"])]

# Calcola IV riga-per-riga con r specifico se presente
iv_list = []
for px, K, T, rr in zip(df["close"].values, df["K"].values, df["T"].values, df["r"].values):
    iv_list.append(implied_vol_call(px, S0, float(K), float(T), float(rr), q))
df = df.assign(iv=np.array(iv_list, dtype=float))
calls = df[np.isfinite(df["iv"])].copy()

# ====== Plot "smirk" ======
uniq_T = np.sort(calls["T"].unique())
if uniq_T.size > 5:
    idx = np.unique(np.clip(np.round(np.linspace(0, uniq_T.size-1, 5)).astype(int), 0, uniq_T.size-1))
    sel_T = uniq_T[idx]
else:
    sel_T = uniq_T

plt.figure(figsize=(8,5))
for T in sel_T:
    dft = calls[calls["T"]==T].sort_values("K")
    if len(dft) < 3:
        continue
    plt.plot(dft["K"].values, dft["iv"].values, marker="o", linestyle="-", label=f"T={T:.3f}y")

plt.xlabel("Strike K"); plt.ylabel("Implied volatility (BlackâScholes)")
plt.title(f"{ticker} IV smirk â {len(sel_T)} maturities")
plt.legend(); plt.grid(True, linewidth=0.4, alpha=0.6)
plt.tight_layout(); plt.show()

# ----- plot IV per T target -----
target_T = np.array([0.103, 0.449, 1.024, 2.270], dtype=float)

uniq_T = np.sort(calls["T"].unique())
picked = []
used_idx = set()
for Ttar in target_T:
    if uniq_T.size == 0: break
    i = int(np.argmin(np.abs(uniq_T - Ttar)))
    while i in used_idx and i+1 < uniq_T.size and abs(uniq_T[i+1]-Ttar) < abs(uniq_T[i]-Ttar):
        i += 1
    used_idx.add(i)
    picked.append((Ttar, uniq_T[i]))

plt.figure(figsize=(8,5))
for Ttar, Tact in picked:
    dft = calls[np.isclose(calls["T"].values, Tact, rtol=0, atol=1e-6)].sort_values("K")
    if len(dft) < 3:
        continue
    plt.plot(dft["K"].values, dft["iv"].values, marker="o", linestyle="-",
             label=f"Tâ{Ttar:.3f}y (actual {Tact:.3f})")

plt.xlabel("Strike K"); plt.ylabel("Implied volatility (BlackâScholes)")
plt.title(f"{ticker} IV smirk â selected maturities")
plt.legend(); plt.grid(True, linewidth=0.4, alpha=0.6)
plt.tight_layout()
plt.savefig("BS_implied_MSTR.png", dpi=200, bbox_inches="tight")
plt.show()
```

# Heston

```python
# ==============================================
#   Heston QMC + LSM @ t_put (soft-call + put)
#   Full-truncation logâEuler, Sobol scrambled
#   Nessun double-discount sul pre-put
# ==============================================
from __future__ import annotations
from dataclasses import dataclass
import os, json
import numpy as np
from numpy.polynomial.hermite import hermval
from scipy.stats import qmc, norm
from scipy.optimize import minimize
from datetime import datetime
```

```python
# -------- base date, notional --------
# Se vuoi legarlo al meta spot_ts, sostituisci questa BASE_DATE.
BASE_DATE = datetime(2025, 10, 20)
N_UNIT = 100.0  # nominale in cassa alla put

def _yf(d: datetime, base: datetime = BASE_DATE) -> float:
    return (d - base).days / 365.0
```

```python
# -------- bond specs (Cr per 100 nominali) --------
def _bond_spec(tipo: str):
    t = str(tipo).strip().upper()
    if t == "2029":
        Cr = 0.14872            # azioni per 100 nominali
        S_star = N_UNIT / Cr    # conversion price
        K_soft = 1.3 * S_star
        return dict(
            Cr=Cr, S_star=S_star, K_soft=K_soft,
            K_put=N_UNIT,  # put in cash = 100
            t_put=_yf(datetime(2028, 6, 1)),
            t_soft_a=_yf(datetime(2026, 12, 2)),
            t_soft_b=_yf(datetime(2029, 11, 29)),
            T=_yf(datetime(2029, 11, 29)),
            name="2029"
        )
    if t in {"2030", "2030B"}:
        Cr = 0.23072
        S_star = N_UNIT / Cr
        K_soft = 1.3 * S_star
        return dict(
            Cr=Cr, S_star=S_star, K_soft=K_soft,
            K_put=N_UNIT,  # put in cash = 100
            t_put=_yf(datetime(2028, 3, 1)),
            t_soft_a=_yf(datetime(2027, 3, 5)),
            t_soft_b=_yf(datetime(2030, 2, 1)),
            T=_yf(datetime(2030, 2, 27)),
            name="2030B"
        )
    raise ValueError("tipo must be '2029' or '2030B'")
```

```python
# -------- NelsonâSiegelâSvensson (OLS) --------
def _nss_basis(m, tau1, tau2):
    m = np.asarray(m, dtype=float)
    def B1(x,t):
        xt = x/t
        den = np.where(xt==0.0, 1.0, xt)
        return (1.0 - np.exp(-xt)) / den
    def B2(x,t):
        xt = x/t
        den = np.where(xt==0.0, 1.0, xt)
        return (1.0 - np.exp(-xt)) / den - np.exp(-xt)
    return B1(m, tau1), B2(m, tau1), B2(m, tau2)

def calibrate_nss_ols(maturities, y):
    maturities = np.asarray(maturities, float)
    y = np.asarray(y, float)

    def obj(z):
        tau1, tau2 = np.exp(z[0]), np.exp(z[1])
        B1, B2, B3 = _nss_basis(maturities, tau1, tau2)
        X = np.column_stack([np.ones_like(maturities), B1, B2, B3])
        beta, *_ = np.linalg.lstsq(X, y, rcond=None)
        r = y - X @ beta
        return float(r @ r)

    z0 = np.log([1.5, 3.0])
    res = minimize(obj, z0, method="Nelder-Mead")
    tau1, tau2 = np.exp(res.x[0]), np.exp(res.x[1])

    B1, B2, B3 = _nss_basis(maturities, tau1, tau2)
    X = np.column_stack([np.ones_like(maturities), B1, B2, B3])
    beta, *_ = np.linalg.lstsq(X, y, rcond=None)

    def nss_yield(m):
        m = np.asarray(m, float)
        b1, b2, b3 = _nss_basis(m, tau1, tau2)
        Xm = np.column_stack([np.ones_like(m), b1, b2, b3])
        return Xm @ beta

    params = {"beta0":beta[0], "beta1":beta[1], "beta2":beta[2], "beta3":beta[3],
              "tau1":tau1, "tau2":tau2}
    return nss_yield, params

# curva NSS di input
yield_maturities = np.array([1/12, 1.5/12, 2/12, 3/12, 4/12, 6/12, 1, 2, 3, 5, 7, 10, 20, 30])
yields = np.array([4.18, 4.15, 4.08, 4.00, 3.95, 3.79, 3.56, 3.46, 3.47, 3.59, 3.78, 4.02, 4.58, 4.60]).astype(float) / 100
nss_yield, _ = calibrate_nss_ols(yield_maturities, yields)

def _yield_at_scalar(m):
    return float(np.asarray(nss_yield(np.array([m], dtype=float))).reshape(-1)[0])
```

```python
# -------- Heston params from JSON (giÃ  sotto Q) --------
hp_path = "mstr_panel_heston_params.json"
if not os.path.exists(hp_path):
    raise FileNotFoundError(f"{hp_path} mancante")
with open(hp_path, "r") as fh:
    hp = json.load(fh)

# Parametri sotto Q direttamente dal file
S0   = float(hp["S0"])
v0   = float(hp["v0"])
kappaQ = float(hp["kappaQ"])
thetaQ = float(hp["thetaQ"])
xi   = float(hp["xi"])
rho  = float(hp["rho"])

# LaTeX line con r e M usati nelle run
M = 2**15
# Per la stampa scegli un r rappresentativo, qui a T=2y
_r_preview = _yield_at_scalar(2.0)
latex_line = (
    r"By a Heston--Q analysis with parameters "
    r"\(\hat{\Theta}=(v_0,\kappa^Q,\theta^Q,\xi,\rho)=("
    f"{v0:.4f},\\,{kappaQ:.4f},\\,{thetaQ:.4f},\\,{xi:.4f},\\,{rho:.4f}"
    r")\), "
    r"risk-free rate \(r=" + f"{_r_preview:.3f}" + r"\), and \(M=2^{{15}}\) quasi--Monte Carlo paths, "
    r"we obtained the following results."
)
print(latex_line)
```

```python
# -------- container risultati --------
@dataclass
class ConvertibleHestonResult:
    price: float
    d1: float; d2: float; d3: float; d4: float
    Sput: np.ndarray; Vput: np.ndarray
    beta: np.ndarray; basis_desc: str
    S_eq_Kp: float

# -------- basi LSM --------
def _hermite_basis(x):
    c0 = [1.0]; c1 = [0.0, 1.0]; c2 = [-1.0, 0.0, 1.0]
    return np.column_stack([hermval(x, c0), hermval(x, c1), hermval(x, c2)])

def _laguerre_basis(y):
    L0 = np.ones_like(y); L1 = 1.0 - y; L2 = 1.0 - 2.0*y + 0.5*y*y
    return np.column_stack([L0, L1, L2])

def _tensor_features(x, y):
    He = _hermite_basis(x); La = _laguerre_basis(y)
    one = He[:, :1]*0 + 1.0
    He1 = He[:, 1:2]; He2 = He[:, 2:3]
    L1  = La[:, 1:2]; L2  = La[:, 2:3]
    return np.column_stack([one, He1, He2, L1, L2, He1*L1, He2*L1])
```

```python
# -------- Sobol -> N(0,1) --------
def _sobol_normals(n_paths, dim, seed=None):
    rng = qmc.Sobol(d=dim, scramble=True, seed=seed)
    U = rng.random_base2(int(np.log2(n_paths))) if (n_paths & (n_paths-1))==0 else rng.random(n_paths)
    U = np.clip(U, 1e-12, 1-1e-12)
    return norm.ppf(U).astype(np.float64)

# -------- simulazione Heston --------
def _simulate_heston_qmc(S0, v0, r, q, kappa, theta, xi, rho, t_grid, n_paths, seed=None):
    dt = np.diff(t_grid); m = dt.size
    S = np.empty((n_paths, m+1)); V = np.empty((n_paths, m+1))
    S[:, 0] = float(S0); V[:, 0] = float(v0)
    Z = _sobol_normals(n_paths, 2*m, seed=seed).reshape(n_paths, m, 2)
    Z1, Z2 = Z[:, :, 0], Z[:, :, 1]
    sqrt_dt = np.sqrt(dt)
    dW1 = Z1 * sqrt_dt
    dW2 = (rho * Z1 + np.sqrt(max(1.0 - rho*rho, 0.0)) * Z2) * sqrt_dt
    mu = (r - q)
    for i in range(m):
        Vp = np.maximum(V[:, i], 0.0)
        V[:, i+1] = np.maximum(V[:, i] + kappa*(theta - Vp)*dt[i] + xi*np.sqrt(Vp)*dW2[:, i], 0.0)
        S[:, i+1] = S[:, i] * np.exp((mu - 0.5*Vp) * dt[i] + np.sqrt(Vp) * dW1[:, i])
    return S, V

def _make_grid(T, m, include_points=()):
    t = np.linspace(0.0, float(T), int(m)+1)
    add = sorted(set([p for p in include_points if 0.0 <= p <= T]))
    if not add: return t
    t_all = np.unique(np.concatenate([t, np.array(add, float)])); t_all.sort(); return t_all
```

```python
def _continuation_after_put(S_tail, t_tail, r, Cr, K_soft, t_b_rel, R):
    n, _ = S_tail.shape
    hit_soft = np.zeros(n, bool); reach_T = np.zeros(n, bool); pv = np.empty(n)
    idx_b = 0 if t_b_rel <= 0 else max(0, int(np.searchsorted(t_tail, min(t_b_rel, t_tail[-1]), "right")) - 1)
    disc = np.exp(-r * t_tail)
    for k in range(n):
        pvk = None
        for j in range(1, idx_b+1):
            if S_tail[k, j] >= K_soft:
                pvk = disc[j] * (Cr * S_tail[k, j]); hit_soft[k] = True; break
        if pvk is None:
            pvk = disc[-1] * max(R, Cr * S_tail[k, -1]); reach_T[k] = True
        pv[k] = pvk
    return pv, hit_soft, reach_T

def _lsm_regression(S_put, V_put_state, Y):
    L = np.log(S_put); x_mu, x_sd = L.mean(), L.std(ddof=0) + 1e-12
    x = (L - x_mu) / x_sd
    Vr = np.sqrt(np.maximum(V_put_state, 0.0))
    y_alpha = max(np.median(Vr), 1e-12); y = Vr / y_alpha
    Phi = _tensor_features(x, y)
    beta, *_ = np.linalg.lstsq(Phi, Y, rcond=None)
    C_hat = Phi @ beta
    return C_hat, beta, "Î¦={1,He1,He2,L1,L2,He1Â·L1,He2Â·L1}; x=std(logS), y=sqrt(v)/Î±", {"x_mu":x_mu,"x_sd":x_sd,"y_alpha":y_alpha}

def _eval_cont_S(beta, S_vals, v_fix, aux):
    L = np.log(S_vals); x = (L - aux["x_mu"]) / aux["x_sd"]
    y = np.full_like(x, np.sqrt(max(v_fix,0.0)) / max(aux["y_alpha"],1e-12))
    Phi = _tensor_features(x, y); return Phi @ beta

def price_convertible_heston_QMC_LSM(
    S0, v0, r, q, kappaQ, thetaQ, xi, rho,
    t0_val, t_a, t_put, t_b, T, K_soft, K_p, R, Cr,
    n_paths, steps_per_year=252, seed=None
) -> ConvertibleHestonResult:

    m_base = int(np.ceil(T * steps_per_year))
    t = _make_grid(T, m_base, include_points=(t_a, t_put, t_b, T))
    idx_a = int(np.searchsorted(t, t_a, "left"))
    idx_put = int(np.searchsorted(t, t_put, "left"))

    S, V = _simulate_heston_qmc(S0, v0, r, q, kappaQ, thetaQ, xi, rho, t, int(n_paths), seed)

    # early soft-call in [t_a, t_put)
    called = np.zeros(n_paths, bool)
    pv_pre_put = np.zeros(n_paths)
    if idx_put > idx_a:
        S_win = S[:, idx_a:idx_put+1]
        hit = (S_win[:, 1:] >= K_soft)
        first_idx = np.where(hit.any(1), hit.argmax(1) + 1, -1)
        for k in range(n_paths):
            j = first_idx[k]
            if j > 0:
                called[k] = True
                tau = t[idx_a + j]
                pv_pre_put[k] = np.exp(-r * tau) * (Cr * S[k, idx_a + j])

    d1 = called.mean()
    alive = ~called
    S_put_all = S[alive, idx_put]
    V_put_all = V[alive, idx_put]

    if S_put_all.size == 0:
        price = pv_pre_put.mean()
        return ConvertibleHestonResult(price, d1, 0.0, 0.0, 0.0,
                                       np.array([]), np.array([]),
                                       np.zeros(7), "n/a", np.nan)

    t_tail = (t[idx_put:] - t[idx_put])
    S_tail = S[alive, idx_put:]
    t_b_rel = max(0.0, t_b - t_put)
    Y_pv_tput, hit_post, reach_T = _continuation_after_put(S_tail, t_tail, r, Cr, K_soft, t_b_rel, R)

    C_hat, beta, basis_desc, aux = _lsm_regression(S_put_all, V_put_all, Y_pv_tput)
    Vput_choice = np.maximum(float(K_p), C_hat)
    put_mask = (Vput_choice <= float(K_p) + 1e-12)  # tolleranza numerica
    cont_mask = ~put_mask

    price = pv_pre_put.mean() + np.exp(-r * max(0.0, t_put - t0_val)) * Vput_choice.mean()

    N = float(n_paths)
    d2 = put_mask.sum() / N
    d3 = (hit_post[cont_mask].sum() if cont_mask.any() else 0) / N
    d4 = (reach_T[cont_mask].sum() if cont_mask.any() else 0) / N

    try:
        v_fix = float(np.median(V_put_all))
        Smin, Smax = float(np.percentile(S_put_all, 1)), float(np.percentile(S_put_all, 99))
        f = lambda s: _eval_cont_S(beta, np.array([s], float), v_fix, aux)[0] - float(K_p)
        a, b = max(1e-8, 0.5*Smin), 2.0*Smax
        fa, fb = f(a), f(b)
        if fa*fb > 0: S_eq = np.nan
        else:
            for _ in range(60):
                m = 0.5*(a+b); fm = f(m)
                if fa*fm <= 0: b, fb = m, fm
                else: a, fa = m, fm
            S_eq = 0.5*(a+b)
    except Exception:
        S_eq = np.nan

    order = np.argsort(S_put_all)
    return ConvertibleHestonResult(
        price=float(price),
        d1=float(d1), d2=float(d2), d3=float(d3), d4=float(d4),
        Sput=S_put_all[order], Vput=Vput_choice[order],
        beta=beta, basis_desc=basis_desc, S_eq_Kp=float(S_eq)
    )
```

```python
# ===== pretty print helper =====
_FEATURES = ["1", "He1", "He2", "L1", "L2", "He1Â·L1", "He2Â·L1"]

def _print_run(label, spec, r, res):
    sep = "â" * 76
    T = float(spec["T"])
    ssum = float(res.d1 + res.d2 + res.d3 + res.d4)
    ok = abs(ssum - 1.0) <= 1e-8
    print(sep)
    print(f"â  {label} | T={T:.2f}y | r(NSS@T)={100*r:.3f}% | Cr={spec['Cr']:.5f} | S*={spec['S_star']:.2f} | K_soft={spec['K_soft']:.2f}")
    print(f"  Prezzo={res.price:.6f}")
    print(f"  d1(pre-call)={res.d1:.6f} | d2(put)={res.d2:.6f} | d3(post-put call)={res.d3:.6f} | d4(reach T)={res.d4:.6f}")
    print(f"  check: d1+d2+d3+d4 = {ssum:.12f}  [{'OK' if ok else 'WARN'}]")
    if np.isfinite(res.S_eq_Kp):
        print(f"  S_eq(K_p)â{res.S_eq_Kp:.6f}")
    print("  Î² LSM:")
    beta = np.asarray(res.beta).reshape(-1)
    for i, b in enumerate(beta):
        name = _FEATURES[i] if i < len(_FEATURES) else f"Î¦{i}"
        print(f"    {name:<7}: {b: .10f}")
    print(sep + "\n")
```
