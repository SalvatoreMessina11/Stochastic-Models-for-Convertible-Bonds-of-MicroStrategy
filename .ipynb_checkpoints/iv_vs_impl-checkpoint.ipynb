## impl vs real

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from pathlib import Path

# --- util ---
def load_series(path: str, date_col: str, value_col: str, name: str):
    p = Path(path)
    if not p.exists():
        return None
    df = pd.read_csv(p, parse_dates=[date_col]).sort_values(date_col)
    return df.set_index(date_col)[value_col].astype(float).rename(name)

# --- carica IV ---
iv = load_series('implied_volatility_5y.csv', 'date', 'Implied Volatility', 'IV')
if iv is None or iv.empty:
    raise RuntimeError("File 'implied_volatility_5y.csv' mancante o vuoto.")

# --- carica RV trailing: preferisci IB, altrimenti custom ---
rv_ib   = load_series('realized_volatility_5y.csv', 'date', 'Realized Volatility', 'RV_trailing')
rv_cust = load_series('custom_volatility_5y.csv',   'date', 'Custom Volatility',   'RV_trailing')

rv_trailing = rv_ib if (rv_ib is not None and not rv_ib.empty) else rv_cust
if rv_trailing is None or rv_trailing.empty:
    raise RuntimeError("N 'realized_volatility_5y.csv' n 'custom_volatility_5y.csv' disponibili.")

print(f"Fonte RV: {'IB HISTORICAL_VOLATILITY' if (rv_ib is not None and not rv_ib.empty) else 'Custom 30d rolling std'}")

# --- RV futura a 30 sedute ---
rv_next30 = rv_trailing.shift(-30).rename('RV_next30')

# --- allinea e pulisci ---
df = pd.concat([iv, rv_next30], axis=1).dropna()
if df.empty:
    raise RuntimeError("Serie allineata vuota dopo dropna. Verifica la sovrapposizione delle date.")

# --- metriche ---
diff = df['RV_next30'] - df['IV']
corr = df['IV'].corr(df['RV_next30'])
bias = diff.mean()
mae  = diff.abs().mean()
mape = (diff.abs() / df['RV_next30']).replace([np.inf, -np.inf], np.nan).dropna().mean()

print(f"Corr(IV, RV_next30): {corr:.3f}")
print(f"Bias (RV_next30 - IV): {bias:.4f}")
print(f"MAE: {mae:.4f}")
print(f"MAPE: {mape:.2%}")

# --- grafico temporale IV(t) vs RV(t+30d) ---
fig_ts = go.Figure()
fig_ts.add_trace(go.Scatter(x=df.index, y=df['IV'], name='Implied Volatility', line=dict(width=2)))
fig_ts.add_trace(go.Scatter(x=df.index, y=df['RV_next30'], name='Realized Vol (next 30d)', line=dict(width=2)))
fig_ts.update_layout(
    title='IV(t) vs Realized Volatility over the Next 30 Trading Days',
    title_x=0.5,
    xaxis_title='Date',
    yaxis_title='Volatility',
    yaxis_tickformat='.1%',
    hovermode='x unified',
    legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01),
    template='plotly_white'
)
fig_ts.show()

# --- scatter di calibrazione IV vs RV_next30 ---
fig_sc = go.Figure()
fig_sc.add_trace(go.Scatter(x=df['IV'], y=df['RV_next30'], mode='markers', name='Points', opacity=0.6))
diag_min = float(df[['IV','RV_next30']].min().min())
diag_max = float(df[['IV','RV_next30']].max().max())
diag_x = np.linspace(diag_min, diag_max, 200)
fig_sc.add_trace(go.Scatter(x=diag_x, y=diag_x, mode='lines', name='y = x'))
fig_sc.update_layout(
    title='Calibration: Implied Volatility vs Next-30-Day Realized Volatility',
    title_x=0.5,
    xaxis_title='IV',
    yaxis_title='RV_next30',
    xaxis_tickformat='.1%',
    yaxis_tickformat='.1%',
    template='plotly_white'
)
fig_sc.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from pathlib import Path

# === aggiorna i CSV da IBKR (IV e prezzi) ===
REFRESH = True            # metti False se vuoi solo leggere i CSV esistenti
SYMBOL  = 'MSTR'
DURATION= '5 Y'
BAR     = '1 day'
IV_CSV  = Path('implied_volatility_5y.csv')
PX_CSV  = Path('prices_5y.csv')

def refresh_from_ibkr():
    try:
        _ = IBKRDataCollector  # verifica presenza classe nel kernel
    except NameError:
        raise RuntimeError("IBKRDataCollector non definita nel kernel. Esegui prima la cella con la classe.")

    app = IBKRDataCollector()
    app.run()
    if not app.connect_to_tws():
        raise RuntimeError("Connessione TWS/Gateway fallita.")

    try:
        data = app.get_historical_volatility_data(
            symbol=SYMBOL, duration_str=DURATION, bar_size_setting=BAR
        )
        # salva IV
        iv = data.get('implied_volatility')
        if iv is None or getattr(iv, 'empty', True):
            raise RuntimeError("IB non ha restituito implied volatility. Verifica abbonamenti dati.")
        iv.to_csv(IV_CSV)  # index=date, col=Implied Volatility

        # salva prezzi
        px_list = data.get('prices', [])
        if not px_list:
            raise RuntimeError("IB non ha restituito prezzi TRADES.")
        pd.DataFrame(px_list)[['date','close']].to_csv(PX_CSV, index=False)

    finally:
        app.disconnect_from_tws()

if REFRESH:
    refresh_from_ibkr()

# --- loader semplice ---
def load_iv(path=IV_CSV):
    if not Path(path).exists():
        raise FileNotFoundError(f"Manca {path}")
    df = pd.read_csv(path, parse_dates=['date']).sort_values('date')
    col = 'Implied Volatility' if 'Implied Volatility' in df.columns else [c for c in df.columns if c != 'date'][0]
    return df.set_index('date')[col].astype(float).rename('iv')

def load_px(path=PX_CSV):
    if not Path(path).exists():
        raise FileNotFoundError(f"Manca {path}")
    df = pd.read_csv(path, parse_dates=['date']).sort_values('date')
    col = 'close' if 'close' in df.columns else [c for c in df.columns if c != 'date'][0]
    return df.set_index('date')[col].astype(float).rename('close')

# === dati necessari ===
iv = load_iv()
px = load_px()

# log staleness
print("IV last:", iv.index.max().date(), "| PX last:", px.index.max().date())

# === serie base ===
r   = px.pct_change().rename('r')
div = iv.diff().rename('d_iv')
df = pd.concat([r, div], axis=1).dropna()

# --- lead/lag opzionale ---
d_iv_lead1 = div.shift(-1).rename('d_iv_lead1')
d_iv_lag1  = div.shift(1).rename('d_iv_lag1')
dd = pd.concat([df, d_iv_lead1, d_iv_lag1], axis=1).dropna()

# === rolling 30 giorni ===
window = 30
roll = pd.DataFrame({
    'date': dd.index,
    'rho_r_dIV_t'  : dd['r'].rolling(window).corr(dd['d_iv']),
    'rho_r_dIV_t+1': dd['r'].rolling(window).corr(dd['d_iv_lead1']),
    'rho_r_dIV_t-1': dd['r'].rolling(window).corr(dd['d_iv_lag1']),
}).dropna()

# === grafico ===
fig = go.Figure()
fig.add_trace(go.Scatter(x=roll['date'], y=roll['rho_r_dIV_t'],
                         mode='lines', name='(r_t, IV_t)', line=dict(width=2)))
fig.add_trace(go.Scatter(x=roll['date'], y=roll['rho_r_dIV_t+1'],
                         mode='lines', name='(r_t, IV_{t+1})', line=dict(width=2, dash='dot')))
fig.add_trace(go.Scatter(x=roll['date'], y=roll['rho_r_dIV_t-1'],
                         mode='lines', name='(r_t, IV_{t-1})', line=dict(width=2, dash='dash')))
fig.add_hline(y=0, line_width=1)
fig.update_layout(
    title='Rolling 30-Day Correlation: Returns vs Changes in Implied Volatility',
    title_x=0.5, xaxis_title='Date', yaxis_title='Correlation ()',
    yaxis_range=[-1, 1], hovermode='x unified',
    legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01),
    template='plotly_white'
)
fig.show()

# --- riepilogo numerico ---
def safe_corr(a, b):
    x = pd.concat([a, b], axis=1).dropna()
    return x.iloc[:,0].corr(x.iloc[:,1])

print("Full-sample correlations")
print(f"(r_t, IV_t)   = {safe_corr(df['r'], df['d_iv']):.3f}")
print(f"(r_t, IV_t+1) = {safe_corr(dd['r'], dd['d_iv_lead1']):.3f}")
print(f"(r_t, IV_t-1) = {safe_corr(dd['r'], dd['d_iv_lag1']):.3f}")
```
